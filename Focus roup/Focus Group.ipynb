```python
import pandas as pd
```

```python
# loading all lines
with open('focus_group_03.txt') as file:
    lines = [line.strip() for line in file]
```

```python
#ignoring certain lines
focus3=[]
for l in lines:
    if(l.startswith("Oksana") or l.startswith("Fred") or l.startswith("Marco") or l.startswith("Mario") or l.startswith("All")):
        continue
    else:
        focus3.append(l)
        
```

```python
# testing
s="Speaker 1 (M) [00:00:40] But my thoughts were to learn about technology now learning more about technology."
print(s.partition("(")[0])
print(s[s.find("(")+1:s.find(")")])
print(s[s.find("[")+1:s.find("]")])
print (s.partition("]")[2])
```

```python
import numpy as np
```

```python
name=[]
gender=[]
time=[]
text=[]
for s in focus3:
    name.append(s.partition("(")[0])
    gender.append(s[s.find("(")+1:s.find(")")])
    time.append(s[s.find("[")+1:s.find("]")])
    text.append(s.partition("]")[2])
```

```python
df3=pd.DataFrame()
```

```python
df3['Name']=np.char.strip(np.array(name))
df3
```

```python
df3['Gender']=np.char.strip(np.array(gender))
df3['Time']=np.char.strip(np.array(time))
df3['Text']=np.char.strip(np.array(text))
df3
```

```python
df3.to_csv('Focus_Group_3_Processed.csv')
```

```python
df3.isnull().sum()
```

```python
#seperate by gender
grouped = df3.groupby(df3['Gender'])
df3_male=(grouped.get_group('M'))
df3_male
```

```python
#reset index
df3_male.reset_index(inplace=True)
df3_male
```

```python
df3_female=(grouped.get_group('F'))
df3_female.reset_index(inplace=True)
df3_female
```

```python
df3_male.to_csv('Focus3 Male.csv')
```

```python
df3_female.to_csv('Focus3 Female.csv')
```
